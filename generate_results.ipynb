```python
from pathlib import Path
import json
from src.data_structures import Instance
from src.solvers import ZeroReductor,BaldoML,BaldoGA,ZeroReductor2
from random import shuffle
import pandas as pd
from src.Gurobi import gurobi,SolverConfig,VAR_TYPE
from IPython.display import clear_output
import os
from matplotlib import pyplot as plt
```

```python
results_path = Path("results")
baldo_data_path = Path("baldo_test_data")
final_results_path = Path("final_results")

```

```python
files = results_path.iterdir()
names_to_use = [file.name for file in files]
shuffle(names_to_use)
```

```python
metric_history = []
metric_history_ml = []
n_items = []
instances = [] 
for file in names_to_use:
    instance = Instance.from_file(baldo_data_path/file)
    instance.file_name = file
    instances.append(instance)
instances.sort(key=lambda x: x.n_items,reverse=False)


```

```python
import cProfile
def main():
    for instance in instances:
        result = open(results_path/instance.file_name,"r")
        result_data = json.load(result)
        result.close()
        instance.result = result_data
        for i in range(1,5):
            threshold = 0.05*i
            sol = ZeroReductor2.solve(instance,pred_threshold=threshold)
            instance.result[f'Z_threshold{round(threshold,2)}_of'] = sol.o
            instance.result[f'Z_threshold{round(threshold,2)}_time'] = sol.time
        final_result_file = open(final_results_path/instance.file_name,"w")
        json.dump(instance.result,final_result_file)
        final_result_file.close()
        print(instance.file_name)
cProfile.run("main()",sort="cumtime")
```
